```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv("balance.txt",delim_whitespace=True)
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity. 
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()`For a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)). 




```python
# Compare the average income based on ethnicity
ave_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print("The average income based on ethnicity is:\n",ave_income_by_ethnicity)

      
```

```python
# Comparing the balance between married vs single people
single_balance = df[df['Married'] == 'No']['Balance'].mean()
married_balance = df[df['Married'] == 'Yes']['Balance'].mean()
print(f'Single people have a higher average balance at {single_balance} vs married people at {married_balance}')
```

```python
# highest income of dataset
highest_income = df["Income"].max()
print('The highest income in the data set:',highest_income)
```

```python
# lowest income of dataset
lowest_income = df["Income"].min()
print('The lowest income in the data set:',lowest_income)
```

```python
# how many cards recorded in data set?
number_of_cards = df.Cards.sum()
print('The number of cards recorded:',number_of_cards)
```

```python
# how many females we have a record for vs males?
num_females = df[df['Gender'] == 'Female']['Gender'].count()
num_males = df[df['Gender'] == 'Male']['Gender'].count()
print('Number of Females:',num_females)
print('Number of Males',num_males)
```
